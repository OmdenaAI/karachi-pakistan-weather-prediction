```python
# Create a scatter plot: Temperature vs. Evapotranspiration
plt.figure(figsize=(10, 6))
plt.scatter(df['temperature_2m_mean'], df['et0_fao_evapotranspiration'], alpha=0.5)
plt.xlabel('Mean Temperature (°C)')
plt.ylabel('Evapotranspiration')
plt.title('Temperature vs. Evapotranspiration')
plt.grid(True)
plt.show()

# Create a scatter plot: Precipitation vs. Evapotranspiration
plt.figure(figsize=(10, 6))
plt.scatter(df['precipitation_hours'], df['et0_fao_evapotranspiration'], alpha=0.5)
plt.xlabel('Precipitation Hours')
plt.ylabel('Evapotranspiration')
plt.title('Precipitation vs. Evapotranspiration')
plt.grid(True)
plt.show()
```

```python
# Calculate correlation with evapotranspiration
corr_with_evapotranspiration = df.corr()['et0_fao_evapotranspiration']
print("Correlation with Evapotranspiration:\n")
print(corr_with_evapotranspiration)
```

```python
import seaborn as sns

# Plot correlation heatmap
correlation_matrix = df[['et0_fao_evapotranspiration', 'precipitation_hours', 'temperature_2m_mean']].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
from scipy.stats import spearmanr

# Calculate Spearman correlation between features and evapotranspiration
spearman_corr_with_et0 = df.drop(['et0_fao_evapotranspiration'], axis=1).apply(lambda x: spearmanr(x, df['et0_fao_evapotranspiration'])[0])

# Print Spearman correlation coefficients
print("Spearman Correlation with Evapotranspiration:\n")
print(spearman_corr_with_et0)
```

```python
import numpy as np

# Calculate Spearman correlation between variables
spearman_corr_matrix, _ = spearmanr(df[['et0_fao_evapotranspiration', 'precipitation_hours', 'temperature_2m_mean']])

# Create a heatmap of Spearman correlation
plt.figure(figsize=(8, 6))
sns.heatmap(spearman_corr_matrix, annot=True, cmap='coolwarm', xticklabels=['Evapotranspiration', 'Precipitation', 'Temperature'], yticklabels=['Evapotranspiration', 'Precipitation', 'Temperature'])
plt.title('Spearman Rank Correlation Heatmap')
plt.show()
```

```python
# Apply transformations
df['log_et0'] = np.log(df['et0_fao_evapotranspiration'])
df['sqrt_precipitation'] = np.sqrt(df['precipitation_hours'])
df['sqrt_temperature'] = np.sqrt(df['temperature_2m_mean'])

# Create scatter plots
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.scatter(df['log_et0'], df['sqrt_precipitation'], alpha=0.5)
plt.xlabel('Log Evapotranspiration')
plt.ylabel('Square Root Precipitation')
plt.title('Log Evapotranspiration vs. Sqrt Precipitation')

plt.subplot(1, 3, 2)
plt.scatter(df['log_et0'], df['sqrt_temperature'], alpha=0.5)
plt.xlabel('Log Evapotranspiration')
plt.ylabel('Square Root Temperature')
plt.title('Log Evapotranspiration vs. Sqrt Temperature')

plt.subplot(1, 3, 3)
plt.scatter(df['sqrt_precipitation'], df['sqrt_temperature'], alpha=0.5)
plt.xlabel('Square Root Precipitation')
plt.ylabel('Square Root Temperature')
plt.title('Sqrt Precipitation vs. Sqrt Temperature')

plt.tight_layout()
plt.show()
```

```python
# Filter data for the city of Karachi
karachi_data = df[df['city'] == 'Karachi']

# Calculate Spearman correlation between features and evapotranspiration for Karachi
spearman_corr_with_et0_karachi = karachi_data.drop(['et0_fao_evapotranspiration'], axis=1).apply(lambda x: spearmanr(x, karachi_data['et0_fao_evapotranspiration'])[0])

# Print Spearman correlation coefficients for Karachi
print("Spearman Correlation with Evapotranspiration for Karachi:\n")
print(spearman_corr_with_et0_karachi)
```

```python
# Select relevant columns for precipitation, temperature, and evapotranspiration
selected_columns = ['precipitation_hours', 'temperature_2m_max','temperature_2m_mean','temperature_2m_min', 'et0_fao_evapotranspiration']
selected_data = karachi_data[selected_columns]

# Calculate Spearman correlation between selected variables for Karachi
spearman_corr_matrix_selected, _ = spearmanr(selected_data)

# Create a heatmap of Spearman correlation for selected variables in Karachi
plt.figure(figsize=(8, 6))
sns.heatmap(spearman_corr_matrix_selected, annot=True, cmap='coolwarm', xticklabels=selected_columns, yticklabels=selected_columns)
plt.title('Spearman Correlation Heatmap for Precipitation, Temperature, and Evapotranspiration (Karachi)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
