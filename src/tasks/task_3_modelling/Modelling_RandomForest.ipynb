```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
import warnings
warnings.filterwarnings('ignore')
```

# Importing dataset

```python
data = pd.read_csv('df_final_Karachi.csv', delimiter=',' )
data.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
Y = data[['precipitation_rate']]
Y
```

```python
data
```

# dropping all the string values

```python
data['time']=pd.to_datetime(data['time'])
data.drop('time', axis=1, inplace=True)
```

```python
data.drop('city', axis=1, inplace=True)
```

```python
exclude_columns =  Y
rest = [col for col in data.columns if col not in exclude_columns]
```

# Extracting training input

```python
X = data[rest]
Y = data[['precipitation_rate']]

print(X)
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=1)
```

# training model by RandomForestRegressor

```python
model=RandomForestRegressor()
model.fit(X_train, Y_train)
```

```python
Y_pred = model.predict(X_test)
Y_pred
```

# Calculating prediction score

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(model, X_train, Y_train, cv=5, scoring='neg_mean_squared_error')
print("Scores:", np.sqrt(-scores))
```

```python
cv_scores = cross_val_score(model, X_train, Y_train, cv=5)
print('Mean Accuracy: ', round(cv_scores.mean()*100, 3), '% Variance: ', round(cv_scores.var()*100, 3),'%')
```
